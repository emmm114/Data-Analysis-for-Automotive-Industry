# LOOKING INTO TATA COMPANY

```python
Tata = cars_engage[cars_engage['Make'] == 'Tata']
#Tata.head()
```

```python
Tata.loc[:,'Ex-Showroom_Price'] = Tata.loc[:,'Ex-Showroom_Price'].str.replace(',', '')
```

```python
Tata.loc[:,'Ex-Showroom_Price'] = Tata.loc[:,'Ex-Showroom_Price'].str.replace('Rs.', '')
```

```python
Tata.loc[:,'Ex-Showroom_Price'] = Tata.loc[:,'Ex-Showroom_Price'].str.replace(' ', '')
```

```python
#Tata.head()
```

```python
Tata['Ex-Showroom_Price'] =  Tata['Ex-Showroom_Price'].astype('int')
```

# Sum of ex-showroom price of each model

```python
Model = Tata.groupby(by = 'Model')['Ex-Showroom_Price'].sum()
Model
```

# Bar graph of the sum of ex-showroom price of each model

```python
bar = Model.plot(kind = "bar")
plt.show()
```

# Pie-chart of cylinders used in a vehicle

```python
Tata['Cylinders'].value_counts().plot(kind = 'pie', autopct = '%1.2f%%')
plt.show()
```

# This shows that there are more 4 cylinder vehicles in Tata company

# Model vs Gears

```python
Tata['Gears'].value_counts().plot(kind = 'bar')
plt.show()
```

# This shows that there are more 5 gear models in Tata Company
